# Forest Cover Classification
In this project, you will use deep learning to predict forest cover type (the most common kind of tree cover) based only on cartographic variables. The actual forest cover type for a given 30 x 30 meter cell was determined from US Forest Service (USFS) Region 2 Resource Information System data. The covertypes are the following:
- Spruce/Fir
- Lodgepole Pine
- Ponderosa Pine
- Cottonwood/Willow
- Aspen
- Douglas-fir
- Krummholz

Independent variables were then derived from data obtained from the US Geological Survey and USFS. The data is raw and has not been scaled or preprocessed for you. It contains binary columns of data for qualitative independent variables such as wilderness areas and soil type.

This study area includes four wilderness areas located in the Roosevelt National Forest of northern Colorado. These areas represent forests with minimal human-caused disturbances, so existing forest cover types are mainly a result of ecological processes rather than forest management practices.

# Preprocess and explore the dataset
Conduct exploratory data analysis to understand the data. See if any data can be removed or balanced. Split dataset for training, validation, and testing.

```python
import numpy as np
import pandas as pd

from collections import Counter

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from tensorflow.keras.utils import to_categorical

dataset = pd.read_csv('cover_data.csv')

#print(dataset.head())
#print(dataset.info())
#print(Counter(dataset['class']))

# labels and features
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 41)

# Standardize Numerical Features
numerical_features = x.select_dtypes(include=['float64', 'int64'])
numerical_columns = numerical_features.columns

# Note: standardizing vs. not standarding one-hot encodings had minimal effect here
ct = ColumnTransformer([('numeric', StandardScaler(), numerical_columns)], remainder='passthrough')
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)

# Convert back to DataFrame
x_train = pd.DataFrame(x_train, columns = x.columns)
x_test = pd.DataFrame(x_test, columns = x.columns)

# Convert label vectors to integers
le = LabelEncoder()
y_train = le.fit_transform(y_train.astype(str))
y_test = le.transform(y_test.astype(str))

y_train = to_categorical(y_train, dtype='int64')
y_test = to_categorical(y_test, dtype='int64')

```

# Build and train model
Regarding hyperparameters, I experimented with:
- batch size
- optimizer type
- learning rate
- hidden units
- number of layers
- dropout parameters
- early stopping
- activation functions

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential(name='cover_type_classification')
model.add(InputLayer(input_shape=(x_train.shape[1],)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(7, activation='softmax'))

# train model
stop = EarlyStopping(monitor='val_loss', mode='min', patience = 1)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs = 20, batch_size = 32, verbose = 1, validation_split = 0.2, callbacks=[stop])
```

# Evaluate model and visualize performance

```python
import matplotlib.pyplot as plt

# evaluate model
loss, acc = model.evaluate(x_test, y_test, verbose = 0)
print("Loss:", loss)
acc_string = "Accuracy: {:.2%}"
print(acc_string.format(acc))

# plot loss (categorical crossentropy) over epochs
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.set_title('model loss')
ax1.set_xlabel('epoch')
ax1.set_ylabel('loss')
ax1.legend(['train', 'validation'], loc='upper left')

# plot accuracy over epochs
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(history.history['accuracy'])
ax1.plot(history.history['val_accuracy'])
ax1.set_title('model accuracy')
ax1.set_xlabel('epoch')
ax1.set_ylabel('accuracy')
ax1.legend(['train', 'validation'], loc='upper left')

plt.show()
```

# Classification Report

```python
from sklearn.metrics import classification_report
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error

# create classification report
y_estimate = model.predict(x_test)
y_estimate = np.argmax(y_estimate, axis = 1)
y_true = np.argmax(y_test, axis = 1)
print("Classification Report:")
print(classification_report(y_true, y_estimate))

# Is the performance reasonable? Yes, we beat the baseline.
dummy_regr = DummyRegressor(strategy='median')
dummy_regr.fit(x_train, y_train)
y_pred = dummy_regr.predict(y_test)
MAE_baseline = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error Baseline:")
print(MAE_baseline)
```
