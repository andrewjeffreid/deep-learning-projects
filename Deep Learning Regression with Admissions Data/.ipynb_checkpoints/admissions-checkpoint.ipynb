# Imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

# Data Preprocessing

```python
# get features and labels
dataset = pd.read_csv('admissions_data.csv')
features = dataset.iloc[:,1:-1]
labels = dataset.iloc[:,-1]

# split data into train and test sets
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size = 0.1, random_state = 42)

# scale data so all columns have equal weight in the learning model
ct = ColumnTransformer([('standardize', StandardScaler(), features.columns)], remainder='passthrough')
features_train_std = ct.fit_transform(features_train)
features_test_std = ct.transform(features_test)

# convert arrays to DataFrame
features_train_set = pd.DataFrame(features_train_std, columns = features_train.columns)
features_test_set = pd.DataFrame(features_test_std, columns = features_test.columns)
```

# Build Model

```python
# use sequential model
model = Sequential(name='admissionpredictor')
num_features = features.shape[1]

# add input layer
input_layer = InputLayer(input_shape = (num_features,))
model.add(input_layer)

# add hidden layer
model.add(Dense(128, activation='relu'))

# add output layer
model.add(Dense(1))

# choose Adam optimizer
opt=Adam(learning_rate = 0.01)
model.compile(loss='mse', metrics=['mae'], optimizer=opt)

# train and evaluate model
history = model.fit(features_train_set, labels_train, epochs = 40, batch_size = 5, validation_split = 0.2, verbose = 1)
val_mse, val_mae = model.evaluate(features_test_set, labels_test, verbose = 0)
print(val_mse, val_mae)
```

# Plot MAE and loss per epoch for training and validation data

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(history.history['mae'])
ax1.plot(history.history['val_mae'])
ax1.set_title('model mae')
ax1.set_ylabel('MAE')
ax1.set_xlabel('epoch')
ax1.legend(['train', 'validation'], loc='upper left')
 
  # Plot loss and val_loss over each epoch
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(history.history['loss'])
ax2.plot(history.history['val_loss'])
ax2.set_title('model loss')
ax2.set_ylabel('loss')
ax2.set_xlabel('epoch')
ax2.legend(['train', 'validation'], loc='upper left')
 
# used to keep plots from overlapping each other  
fig.tight_layout()
```
